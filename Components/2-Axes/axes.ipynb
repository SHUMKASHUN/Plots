```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
plt.style.use('./axes.mplstyle')
# create axes
fig, axs = plt.subplots(1, 2,layout='constrained') 
np.random.seed(19680801)
t = np.arange(100)
x = np.cumsum(np.random.randn(100))
# using ax[i] to plot on each axes
lines = axs[0].plot(t, x)
```

```python
# Axes scale
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sinc(x)

fig, ax = plt.subplots()
ax.plot(x, y)
# the default margin is 0.05
ax.margins(0.2, 0.5)
```
