```python

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
plt.style.use('./ticks.mplstyle')

fig, ax = plt.subplots(figsize=(4, 2.5))
ax.plot(np.arange(10))

# customize the tick label 
# It is possible to customize the ticks and tick labels with either high-level methods like set_xticks or set the locators and formatters directly on the axis.

ax.set_yticks([2,4,6,8], labels=['$2', '$4', '$6', '$8'])
ax.set_xticks([0,6,8], labels=['$x0', '$x6', '$x8'])
```

```python
fig, ax = plt.subplots(figsize=(4, 2.5))
ax.plot(np.arange(10))
# Style ticks
# you can choose what you want to customize from the style sheet.
# e.g. show ticks and label on the top
# Speciy axis to {x,y,both}

ax.tick_params(axis='x', top=True, labeltop=True, color='red', 
               labelcolor='green',direction="in")
ax.tick_params(axis='y', right=True, labelright=True, color='red')

#It also can control the tick labels:
# labelsize (fontsize)
# labelcolor (color of the label)
# labelrotation
# labelbottom, labeltop, labelleft, labelright
ax.tick_params(axis='x', labelsize=10, labelrotation=45, labelcolor='blue', pad=20)
```
