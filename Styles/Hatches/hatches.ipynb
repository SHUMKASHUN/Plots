```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plt.style.use('./hatches.mplstyle')

fig, ax = plt.subplots(1, 1,layout='constrained') 

e1 = mpl.patches.Ellipse((3, 3), width=1, height=1, fill=False, zorder=2, hatch='/')

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
ax.add_patch(e1)
ax.plot(x, y)
```
