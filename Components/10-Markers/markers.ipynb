```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plt.style.use('./markers.mplstyle')

fig, ax = plt.subplots(1, 1,layout='constrained') 

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

ax.plot(x, y, marker='o', markersize=7, markerfacecolor='white', markeredgewidth=2, markeredgecolor='black')
```
