```python

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
plt.style.use('./legend.mplstyle')
# set the extent on the axes 
# default
fig, ax = plt.subplots(figsize=(4, 2.5), layout='constrained')
ax.plot([1, 2, 3], label='Inline label')
ax.legend()
```

```python
# you can also directly customize the legend by passing  handles and labels
fig, ax = plt.subplots(figsize=(4, 2.5), layout='constrained')
line_up, = ax.plot([1, 2, 3], label='Line 2')
line_down, = ax.plot([3, 2, 1], label='Line 1')
ax.legend([line_up, line_down], ['Line Up', 'Line Down'])

```

```python
# Legend handles don't have to exist on the Figure or Axes in order to be used.
import matplotlib.patches as mpatches
import matplotlib.lines as mlines

fig, ax = plt.subplots(figsize=(4, 2.5), layout='constrained')
red_patch = mpatches.Patch(color='red', label='Line New')
green_line = mlines.Line2D([], [], color='green', label='Line Green')
line_up, = ax.plot([1, 2, 3], label='Line 2')
line_down, = ax.plot([3, 2, 1], label='Line 1')
ax.legend([line_up, line_down, red_patch, green_line], ['Line Up', 'Line Down', 'Line New', 'Line Green'])
```

```python
# Legend position
fig, ax = plt.subplots(figsize=(4, 2.5))
red_patch = mpatches.Patch(color='red', label='Line New')
green_line = mlines.Line2D([], [], color='green', label='Line Green')
line_up, = ax.plot([1, 2, 3], label='Line 2')
line_down, = ax.plot([3, 2, 1], label='Line 1')
ax.legend(loc="upper left",handles=[line_up, line_down, red_patch, green_line], labels=['Line Up', 'Line Down', 'Line New', 'Line Green'])


# (bbox_to_anchor=(0., 1.02, 1., .102),mode="expand",bbox_transform=fig.transFigure)
```

```python
# Sometimes it makes more sense to place a legend relative to the (sub)figure rather than individual Axes.
fig, ax = plt.subplots(figsize=(4, 2.5), layout='constrained')
red_patch = mpatches.Patch(color='red', label='Line New')
green_line = mlines.Line2D([], [], color='green', label='Line Green')
line_up, = ax.plot([1, 2, 3], label='Line 2')
line_down, = ax.plot([3, 2, 1], label='Line 1')
fig.legend(loc="outside lower center",ncol=2,frameon=False, bbox_to_anchor=(0.5, -0.25), handles=[line_up, line_down, red_patch, green_line], labels=['Line Up', 'Line Down', 'Line New', 'Line Green'])

#If you want to put more legend, you need to use ax.add_artist(first_legend) to manually add the first legend
```
