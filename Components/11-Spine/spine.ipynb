```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plt.style.use('./spine.mplstyle')

fig, ax = plt.subplots(1, 1,layout='constrained') 

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# more apis are available at https://matplotlib.org/stable/api/spines_api.html
ax.spines['left'].set(position=('outward', 10),linewidth=2.5, alpha=0.7, color="red", bounds=(4,6))
ax.spines['bottom'].set(position=('outward', 10),linewidth=2.5) 
ax.spines['right'].set(visible=False)
ax.plot(x, y)
```
